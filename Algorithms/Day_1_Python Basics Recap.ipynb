### ğŸ“Œ 4. Beginner-Friendly Examples
 O(1) â€“ Constant Time

```python
def get_first_element(arr):
    return arr[0]  # Always 1 step, no matter the array size

print(get_first_element([10, 20, 30]))
```

  O(n) â€“ Linear Time 

```python
def print_all(arr):
    for item in arr:
        print(item)  # Executes n times

print_all([1, 2, 3, 4])
```

O(nÂ²) â€“ Quadratic Time





```python
def print_pairs(arr):
    for i in arr:
        for j in arr:
            print(i, j)  # Executes n*n times

print_pairs([1, 2, 3])
```

O(log n) â€“ Logarithmic Time





```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return True
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return False

print(binary_search([1, 2, 3, 4, 5], 4))
```

 O(2â¿) â€“ Exponential Time





```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(5))  # Slow for big n
```

### ğŸ“Œ 5. Space Complexity Example





```python
def sum_list(arr):
    total = 0  # O(1) space
    for num in arr:  # No extra space based on n
        total += num
    return total

print(sum_list([1, 2, 3, 4]))
```
